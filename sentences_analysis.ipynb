```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
eng = pd.read_table('englishaa',names=['eng'])
de = pd.read_table('germanaa',names=['de'])
eng.head()
```

```python
de.head()
```

```python
plt.figure(figsize=(7,6))
lengths = [len(x.split()) for x in eng.eng]
plt.hist(lengths,color="#3F5D7D", bins=15)
plt.title('Histogram of lengths of English sentences')
plt.show()
```

```python
plt.figure(figsize=(7,6))
plt.boxplot(lengths)
plt.show()
```

```python
plt.figure(figsize=(7,6))
lengths = [len(x.split()) for x in de.de]
plt.hist(lengths,color="#3F5D7D", bins=15)
plt.title('Histogram of lengths of German sentences')
plt.show()
```

```python
plt.figure(figsize=(7,6))
plt.boxplot(lengths)
plt.show()
```

```python
import nltk
from nltk.corpus import stopwords

w = []
for x in eng.eng[:2400]:
    w += x.split() 
words = [word for word in w if word.lower() not in stopwords.words() and word.isalpha() ]
freqdist = nltk.FreqDist(words)
plt.figure(figsize=(16,5))
freqdist.plot(50)
```
